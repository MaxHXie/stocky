## Machine Learning time

### Create labels for your companies

Here we create labels for all companies that basically tracks the change in closing price over the last "hm_days", which then could be used as a feature for e.g. machine learning.

```python
def process_data_for_labels(name):
    hm_days = 7
    df = pd.read_csv('joined_closing_prices.csv', index_col=0)
    names = df.columns.values.tolist()
    df.fillna(0, inplace=True)

    for i in range(1, hm_days+1):
        df['{}_{}d'.format(name, i)] = (df[name].shift(-i) - df[name]) / df[name]

    df.fillna(0, inplace=True)
    return names, df
```

### Create machine learning target function

There are three target labels buy, sell, hold.

If the stock price has increased by "requirement" percent for the last 7 days if will will send a buy signal. If it has decreased then it will send a sell signal, otherwise it will send a hold signal.

```python
def buy_sell_hold(*args):
    cols = [c for c in args]
    requirement = 0.025
    for col in cols:
        if col > requirement:
            return 1
        if col < -requirement:
            return -1
    return 0
```

```python
def extract_featureset(name):
    names, df = process_data_for_labels(name)
    df['{}_target'.format(name)] = list(map(buy_sell_hold,
                                             df['{}_1d'.format(name, i)],
                                             df['{}_2d'.format(name, i)],
                                             df['{}_3d'.format(name, i)],
                                             df['{}_4d'.format(name, i)],
                                             df['{}_5d'.format(name, i)],
                                             df['{}_6d'.format(name, i)],
                                             df['{}_7d'.format(name, i)],
                                            ))
    vals = df['{}_target'.format(name)].values.tolist()
    str_vals = [str(i) for i in vals]
    print('Data spread:', Counter(str_vals))
    
    df.fillna(0, inplace=True)
    df = df.replace([np.inf, -np.inf], np.nan)
    df.dropna(inplace=True)
    
    #Normalize all data
    df_vals = df[[name for name in names]].pct_change()
    df_vals = df_vals.replace([np.inf, -np.inf], 0)
    df_vals.fillna(0, inplace=True)
    
    X = df_vals.values
    y = df['{}_target'.format(name)].values
    
    return X, y, df

extract_featureset("ABB")
```

```python
def do_ml(name):
    X, y, df = extract_featureset(name)
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
    
    #clf = neighbors.KNeighborsClassifier()
    clf = VotingClassifier([('lsvc', svm.LinearSVC()),
                            ('knn', neighbors.KNeighborsClassifier()),
                            ('rfor', RandomForestClassifier())])
    
    clf.fit(X_train, y_train)
    confidence = clf.score(X_test, y_test)
    predictions = clf.predict(X_test)
    print('Predicted spread:', Counter(predictions))
    
    return confidence

do_ml("ABB")
```
